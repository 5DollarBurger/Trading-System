# Trading System

## Market data engine
- This module should mock market data for symbols ‘AAPL’, ‘TSLA’ and ‘GOOG’.
- Each symbol’s market data should be within a prespecified band. This band should be configurable.
- Market data should have 3 values. Best Bid, Best Offer and Last trade price.

## Order placement engine
- Users must be able to place orders via this module.
- User should be able to provide the symbol name, order type (limit or market), order qty and the price if limit order.
- All orders should be written to a db including the timestamp.

## Matching engine
- This module is responsible for matching orders of users.
- Orders should be matched on a price – time priority basis. You can find enough articles in internet on how this simple matching algorithm works if you need any further information.
- All trades should be written to a db including the timestamp.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from random import uniform
from datetime import datetime
import sqlite3
import os
```

## Section 1: market data engine
The goal of this class is to randomly generate the initial state of the market at its opening, before the trades come in. The client calling this object expects to receive empty order books for each stock, and their respective last traded prices. This output instance is henceforth referred to as the $\textit{market state}$. The client must be able to set the random generation parameters, namely the upper and lower limits of each stock's last traded price.

Assumptions:
- opening prices are uniformly distributed within specified ranges
- no orders have been placed for stocks at market opening

```python
class MarketGen:
    """
    Random generation of the market state. Books are empty at initial state.
    
    Attributes:
        last (dict): Dictionary of last matched prices.
        buy_aapl (list(dict)): List of buy limit orders for AAPL.
        buy_goog (list(dict)): List of buy limit orders for GOOG.
        buy_tsla (list(dict)): List of buy limit orders for TSLA.
        sell_aapl (list(dict)): List of sell limit orders for AAPL.
        sell_goog (list(dict)): List of sell limit orders for GOOG.
        sell_tsla (list(dict)): List of sell limit orders for TSLA.
        pairs (list(dict)): List of cleared order pairs.
        trades (DataFrame): Cleared market orders.
        orders (list(dict)): List of all orders.
        
    """
    
    def __init__(self, aapl_range=[310.0, 325.0], goog_range = [1433.9, 1526.69], tsla_range=[547.2, 917.42]):
        """
        Args:
            aapl_range list(float): Band of initial AAPL price.
            goog_range list(float): Band of initial GOOG price.
            tsla_range list(float): Band of initial TSLA price.
        
        """
        # Randomly generate opening prices for each stock
        self.last = dict(AAPL=[uniform(aapl_range[0], aapl_range[1])], 
                    GOOG=[uniform(goog_range[0], goog_range[1])],
                    TSLA=[uniform(tsla_range[0], tsla_range[1])])
        
        # Initiate empty order books for each stock
        self.buy_aapl = []
        self.buy_goog = []
        self.buy_tsla = []
        self.sell_aapl = []
        self.sell_goog = []
        self.sell_tsla = []
        self.pairs = [] # list of cleared market - counterparty pairs
        
        self.trades = pd.DataFrame()
        try:
            os.remove("trades.db")
        except:
            pass
        conn_trades = sqlite3.connect("trades.db")
        c_trades = conn_trades.cursor()
        c_trades.execute("CREATE TABLE TRADES (Time TIMESTAMP, Symbol CHAR (4), Price FLOAT, Volume INT)")
        conn_trades.commit()
        
        self.orders = []
        try:
            os.remove("orders.db")
        except:
            pass
        conn_orders = sqlite3.connect("orders.db")
        c_orders = conn_orders.cursor()
        c_orders.execute("CREATE TABLE ORDERS (Time TIMESTAMP, Symbol CHAR (4), Price FLOAT, \
                         Market BOOL, Buy BOOL, Volume INT)")
        conn_orders.commit()
```

## Section 2: matching engine
The goal of this class is to update the order books at every order placement. When a limit order is placed, the order is appended, queued first by price and second by the order's timestamp. When a market order is placed, counterparty order(s) are cleared according to queued priority. The primary output is hence the updated market state, while the secondary output is a historical database of cleared histories.

```python
class Match:
    """
    Performs matching operations at each order placement.
    
    """
    
    def __init__(self, MSInst, order):
        """
        Args:
            MSInst (object): Current market state.
            order (dict): order parameters.
        
        """
        
        # Pull stock's order book
        if order["tick"]=="AAPL":
            buy_tick = MSInst.buy_aapl
            sell_tick = MSInst.sell_aapl
            
        elif order["tick"]=="GOOG":
            buy_tick = MSInst.buy_goog
            sell_tick = MSInst.sell_goog
            
        elif order["tick"]=="TSLA":
            buy_tick = MSInst.buy_tsla
            sell_tick = MSInst.sell_tsla
            
        else:
            print("Please enter valid ticker symbol.")
            return None
        
        # market orders
        if order["market"]==True:
            
            # market buy orders
            if order["buy"]==True:
                
                # check for availability
                if len(sell_tick)==0:
                    print("No volume in market.")
                    return None
                
                elif sum(l["volume"] for l in sell_tick)<order["volume"]:
                    print("Not enough volume in market.")
                    return None
                
                # perform matching
                matched = self.matching(sell_tick, order["volume"])
                sell_tick = matched[0]

            # market sell orders
            else:
                
                # check for availability
                if len(buy_tick)==0:
                    print("No volume in market.")
                    return None
                
                elif sum(l["volume"] for l in buy_tick)<order["volume"]:
                    print("Not enough volume in market.")
                    return None

                # perform matching
                matched = self.matching(buy_tick, order["volume"])
                buy_tick = matched[0]
                
            # log trade
            MSInst.pairs.append((order, matched[1]))
            tmp = order.copy()
            del tmp["market"] 
            MSInst.trades = MarketState.trades.append(tmp, ignore_index=True)
            MSInst.last[order["tick"]].append(matched[1][-1]["price"])

        # limit buy orders
        elif order["market"]==False:
            
            # limit buy orders
            if order["buy"]==True:
                buy_tick = self.limitbuy(buy_tick, order)

            # limit sell orders
            else:
                sell_tick = self.limitsell(sell_tick, order)
        
        # update respective order books
        if order["tick"]=="AAPL":
            MSInst.buy_aapl = buy_tick
            MSInst.sell_aapl = sell_tick
            
        elif order["tick"]=="GOOG":
            MSInst.buy_goog = buy_tick
            MSInst.sell_goog = sell_tick
            
        else:
            MSInst.buy_tsla = buy_tick
            MSInst.sell_tsla = sell_tick
    
    def matching(self, ob, volume):
        """
        Matches a counterparty for a market order, clears counterparty from order book.

        Args:
            ob (list(dict)): Order book.
            volume (int): Number of shares in market order.

        Returns:
            tuple: Cleared order book, counterparties.

        """
        
        # find counterparty(s)
        filled_total = 0
        volume_left = volume
        counterparties = []
        
        while filled_total < volume:
            
            # number of shares filled by this counterparty
            filled = min(ob[0]["volume"], volume_left)
            counterparty = ob[0].copy()
            counterparty["volume"] = filled
            
            # clear trade
            ob[0]["volume"] = ob[0]["volume"] - filled
            if ob[0]["volume"] == 0:
                ob.pop(0)
                
            # record counterparty
            counterparties.append(counterparty)
            filled_total += filled
            volume_left -= filled
        

        return ob, counterparties;
    
    def limitbuy(self, buy_OB, order):
        """
        Appends buy order book with a new limit buy, prioritising the order on decending price 
        before ascending time.
        
        Args:
            buy_OB (list(dict)): Buy order book.
            order (dict): Parameters of limit buy order.
            
        Returns:
            dict: Updated order book.
            
        """
        # add order to order book
        buy_OB.append({k: order[k] for k in ["time", "tick", "price", "volume"]})
        
        # prioritise highest price first, earliest order first
        buy_OB = sorted(buy_OB, key = lambda i: (-i["price"], i["time"]))
        
        return buy_OB
    
    def limitsell(self, sell_OB, order):
        """
        Appends buy order book with a new limit sell, prioritising the order on ascending price 
        before ascending time.
        
        Args:
            sell_OB (list(dict)): Sell order book.
            order (dict): Parameters of limit Sell order.
            
        Returns:
            dict: Updated order book.
            
        """
        # add order to order book
        sell_OB.append({k: order[k] for k in ["time", "tick", "price", "volume"]})
        
        # prioritise highest price first, earliest order first
        sell_OB = sorted(sell_OB, key = lambda i: (i["price"], i["time"]))
        
        return sell_OB
```

## Section 3: order placement engine
The primary purpose of this class is to process a client's order, before delivering an order instance to the maching engine. The secondary objective is to store all orders within a database. Order parameters to be processed are as follows:
- stock ticker
- type: buy or sell
- market or limit
- price level (for limit)
- number of shares
- timestamp of order

```python
class PlaceOrd():
    """
    Processes client's order parameters.
    
    Attributes:
        df_tsla_sell
        
    """
    
    def __init__(self, MSInst, **order_params):
        """
        Args:
            MSInst (object): Current market state.
            tick (str): Stock's ticker symbol.
            buy (bool): True for buy, False for sell.
            market (bool): True for market order, False for limit order.
            price (float): Purchase price.
            volume (int): Number of shares.
            
        """
        # store order parameters in a dictionary
        order = order_params
        order["time"] = datetime.now().strftime('%Y-%m-%d%H-%M-%S-%f')# + str(uuid4())
        
        # append order dictionary to orders
        MSInst.orders.append(order)
        MSInst.df_orders = pd.DataFrame(MarketState.orders, columns=['time', 'tick', 'price', \
                                                                     'market', 'buy', 'volume'])
        
        # process new order
        Match(MSInst, order)
        
        # save updated order books to dataframes
        if order["tick"]=="AAPL":
            MSInst.df_aapl_buy = pd.DataFrame(MarketState.buy_aapl, columns=['price', 'time', 'volume'])
            MSInst.df_aapl_sell = pd.DataFrame(MarketState.sell_aapl, columns=['price', 'time', 'volume'])
            
        elif order["tick"]=="GOOG":
            MSInst.df_goog_buy = pd.DataFrame(MarketState.buy_goog, columns=['price', 'time', 'volume'])
            MSInst.df_goog_sell = pd.DataFrame(MarketState.sell_goog, columns=['price', 'time', 'volume'])
            
        elif order["tick"]=="TSLA":
            MSInst.df_tsla_buy = pd.DataFrame(MarketState.buy_tsla, columns=['price', 'time', 'volume'])
            MSInst.df_tsla_sell = pd.DataFrame(MarketState.sell_tsla, columns=['price', 'time', 'volume'])
        
        # save orders and trades to db
        conn_orders = sqlite3.connect("orders.db")
        MSInst.df_orders.to_sql('ORDERS', conn_orders, if_exists='replace')
        
        conn_trades = sqlite3.connect("trades.db")
        MSInst.trades.to_sql('TRADES', conn_trades, if_exists='replace')
```

## Testing the code for a sequence of trades
TSLA trade sequence:
1. Market opens with randomised prices
2. First limit sell order on 10 shares
3. 1 microsecond later, a second limit sell order on 5 shares (best offer)
4. Market buy order for 6 shares
5. First limit buy order on 2 shares
6. 1 microsecond later, a second limit buy order on 10 shares at the same price level

```python
# Step 1: inititate market
MarketState = MarketGen(aapl_range=[310.0, 325.0], goog_range = [1433.9, 1526.69], tsla_range=[547.2, 917.42])
print(MarketState.last)
```

```python
print("--- MARKET BUY TEST ---")

# Step 2: first limit sell order
PlaceOrd(MarketState, tick='TSLA', price=MarketState.last['TSLA'][-1]*1.02, volume=10, buy=False, market=False)

import time
time.sleep(0.001)

# Step 3: second limit sell order
POI = PlaceOrd(MarketState, tick='TSLA', price=MarketState.last['TSLA'][-1]*1.01, volume=5, buy=False, market=False)
print("Sell order book before matching:")
print(MarketState.df_tsla_sell)

# Step 4: market buy order
PlaceOrd(MarketState, tick='TSLA', volume=6, buy=True, market=True)
print("Updated sell order book:")
print(MarketState.df_tsla_sell)

# trade record
print("Trade book:")
print(MarketState.trades)

# Visualise price movements across trades
plt.plot(MarketState.last["TSLA"])
plt.title("Price movement")
plt.xlabel("Trade sequence")
plt.ylabel("Last traded price")
```
